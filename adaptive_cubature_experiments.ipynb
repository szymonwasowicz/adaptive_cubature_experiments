### Adaptive Cubature Experiments (CPU)
This notebook demonstrates numerical integration over simplices using adaptive subdivision logic.

```python
from adaptive_cubature_experiments import *
```

#### 2D Experiment: Equilateral Triangle
Integrating a specific function over an equilateral triangle with visualization of the adaptive mesh.

```python
f = lambda x: 1 / (2 * (x[:, 1] + np.sqrt(3)))
S = np.array([[-1, np.sqrt(3)], [0, 0], [1, np.sqrt(3)]])
epsilon = 0.00001
exact_integral = 1-np.log(2)

experiment_2dim_plot(f, S, epsilon, exact_integral)
```

#### 2D Experiment: Unit Simplex
Running the optimized 2D experiment on a standard unit simplex.

```python
f = lambda x: 1 / (1 + np.sum(x, axis=-1))
S = np.vstack([np.zeros(2), np.eye(2)])
epsilon = 0.0000001
exact_integral = 1-np.log(2)

experiment_2dim(f, S, epsilon, exact_integral)
```

#### 3D Experiment: General n-dimensional Logic
Testing the `experiment_ndim` function on a 3-dimensional unit simplex.

```python
# Experiment on 3-dimensional unit simplex

f = lambda x: 1 / (1 + np.sum(x, axis=-1))
S = np.vstack([np.zeros(3), np.eye(3)])
epsilon = 0.0001
exact_integral = 1/2*np.log(2) - 1/4

experiment_ndim(f, S, epsilon, exact_integral)
```
